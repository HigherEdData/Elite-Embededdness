## Make School-level Forbes 400 Alumni By Sector Counts

### For Table 4 of Elite Embeddedness

By Charlie Eaton and Albina Gibadullina

```stata
use d_forbes400_alum_board_1989_2003_2017, clear
```

```stata
set more off

use id boardunitid1 boardunitid2 mbaunitid collegeunitid  ///
boardtop30private boardtop30public pehedge otherfin technology worth inherited ///
collegeschooltype mbaschooltype year using d_forbes400_alum_board_1989_2003_2017, clear
keep if year==2003 | year==2017
drop if boardunitid1==.

collapse (count) forbesboard1=id (sum) otherboard1=other inheritedboard1=inherited pehedgeboard1=pehedge otherfinboard1=otherfin ///
technologyboard1=technology, by (boardunitid1 year)
sum pehedge otherfin technology forbesboard
rename boardunitid1 unitid
save d_board1_forbes_400_count, replace

use id boardunitid1 boardunitid2 mbaunitid collegeunitid  ///
boardtop30private boardtop30public pehedge otherfin technology worth inherited ///
collegeschooltype mbaschooltype year using d_forbes400_alum_board_1989_2003_2017, clear
keep if year==2003 | year==2017
drop if boardunitid2==.

collapse (count) forbesboard2=id (sum) otherboard2=other inheritedboard2=inherited pehedgeboard2=pehedge otherfinboard2=otherfin ///
technologyboard2=technology, by (boardunitid2 year)
sum pehedge otherfin tech forbesboard
rename boardunitid2 unitid
save d_board2_forbes_400_count, replace

use id boardunitid1 boardunitid2 mbaunitid collegeunitid  ///
boardtop30private boardtop30public pehedge otherfin technology worth inherited ///
collegeschooltype mbaschooltype year using d_forbes400_alum_board_1989_2003_2017, clear
keep if year==2003 | year==2017
drop if mbaunitid==.

gen other=0
replace other=1 if pehedge!=1 & otherfin!=1 & technology!=1
*foreach var in pehedge otherfin technology other {
*    replace `var'=0 if inherited==1
*}

collapse (count)forbes_mba=id (sum) other_mba=other inherited_mba=inherited pehedge_mba=pehedge otherfin_mba=otherfin ///
technology_mba=technology, by (mbaunitid year)
sum pehedge otherfin tech forbes
rename mbaunitid unitid
save d_mba_forbes_400_count, replace

use id boardunitid1 boardunitid2 mbaunitid collegeunitid  ///
boardtop30private boardtop30public pehedge otherfin technology worth inherited ///
collegeschooltype mbaschooltype year using d_forbes400_alum_board_1989_2003_2017, clear
keep if year==2003 | year==2017
drop if collegeunitid==.

gen other=0
replace other=1 if pehedge!=1 & otherfin!=1 & technology!=1
*foreach var in pehedge otherfin technology other {
*    replace `var'=0 if inherited==1
*}

collapse (count)forbes_ba=id (sum) other_ba=other inherited_ba=inherited pehedge_ba=pehedge otherfin_ba=otherfin ///
technology_ba=technology, by (collegeunitid year)
sum pehedge otherfin tech forbes
rename collegeunitid unitid
save d_ba_forbes_400_count, replace

use id boardunitid1 boardunitid2 mbaunitid collegeunitid jdunitid  ///
boardtop30private boardtop30public pehedge otherfin technology worth inherited ///
collegeschooltype mbaschooltype year using d_forbes400_alum_board_1989_2003_2017, clear
keep if year==2003 | year==2017
drop if jdunitid==.

gen other=0
replace other=1 if pehedge!=1 & otherfin!=1 & technology!=1
*foreach var in pehedge otherfin technology other {
*    replace `var'=0 if inherited==1
*}

collapse (count)forbes_jd=id (sum) other_jd=other inherited_jd=inherited pehedge_jd=pehedge otherfin_jd=otherfin ///
technology_jd=technology, by (jdunitid year)
sum pehedge otherfin tech forbes
rename jdunitid unitid
save d_jd_forbes_400_count, replace
```
