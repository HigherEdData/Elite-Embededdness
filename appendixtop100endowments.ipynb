```stata
quietly {
foreach year in 18 {
local min1=`year'-1
copy https://nces.ed.gov/ipeds/datacenter/data/F`min1'`year'_F2_Data_Stata.zip F`min1'`year'_F2_Data_Stata.zip, replace
unzipfile F`min1'`year'_F2_Data_Stata, replace
insheet  using f`min1'`year'_f2_data_stata.csv, clear
keep unitid f2h02
    gen year="20"+"`year'"
destring year, replace
save privateendowment`year', replace
}

foreach year in 18 {
local min1=`year'-1
 quietly copy https://nces.ed.gov/ipeds/datacenter/data/F`min1'`year'_F1A_Data_Stata.zip F`min1'`year'_F1A_Data_Stata.zip, replace
 unzipfile F`min1'`year'_F1A_Data_Stata, replace
 insheet  using f`min1'`year'_f1a_data_stata.csv, clear
 keep unitid f1h02
    gen year="20"+"`year'"
 destring year, replace
save publicendowment`year', replace
}

foreach year in 2017 {
local min1=`year'-1
  copy https://nces.ed.gov/ipeds/datacenter/data/HD`year'_Data_Stata.zip HD`year'_Data_Stata.zip, replace
 unzipfile HD`year'_Data_Stata, replace
insheet  using hd`year'_data_stata.csv, clear
keep control unitid instnm
    gen year=`year'+1
save hd`year', replace
}
}
merge 1:1 unitid year using privateendowment18, gen(mergepriv)
merge 1:1 unitid year using publicendowment18, gen(mergepub)
```

```stata
gen endowbillions=f1h02 / 1000000000
replace endowbillions=f2h02 / 1000000000 if endowbillions==.
gsort -endowbillions
label define control 1 "public" 2 "private"
label values control control
label var endowbillions "Endowment in billons"
list instnm control endowbillions if _n<10
```
