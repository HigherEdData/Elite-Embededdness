```stata
clear all
cd "/Users/Charlie/Dropbox/Bankers in the Ivory Tower Book/2 Our New Financial Oligarchy/Data/Boards 2000 to 2017"

use board_endow_donation_directory_20181221, clear
sort instnm
drop if top30private!=1 & top30public!=1
drop if pctnonfinance==.
drop if year>1999 & year<2003
replace pctfinall=pctfinall*100
```

```stata
* Brown unitid 217156 missing data before 2008, Brandeis unitid 165015 before 2004 if unitid==217156
```

```stata
gen pctalt1989=pctalt if minyear==1989
gen pctfinall1989=pctfinall if minyear==1989
gen pctaltpublic=pctalt if top30public==1
gen pctfinallpublic=pctfinall if top30public==1
gen pctaltprivate=pctalt if top30private==1
gen pctfinallprivate=pctfinall if top30private==1

collapse (mean) pctalt1989 pctfinall1989 pctaltpublic pctfinallpublic pctaltprivate pctfinallprivate, by(year)
list
```

```stata
label var pctalt1989 "Top 22 private"
label var pctfinall1989 "Top 22 private" 
label var pctaltpublic "Top 30 public" 
label var pctfinallpublic "Top 30 public" 
label var pctaltprivate "Top 30 private" 
label var pctfinallprivate "Top 30 private"
```

```stata
set scheme plotplain
tw (connect pctalt1989 pctaltpublic pctaltprivate year), ///
xlabel(1985(5)2020) scale(1.4) xtitle("") ///
ytitle("% from PE & hedge funds") legend(size(medium) order(1 3 2)) ///
ylabel(0(10)40) legend(off) name(pehedge, replace) ///
title(% from PE and hedge funds)
```

```stata
tw (connect pctfinall1989 pctfinallpublic pctfinallprivate year), ///
xlabel(1985(5)2020,) xtitle("") scale(1.4) ///
ytitle("% from all of finance") ///
title("% from all of finance") legend(off) name(finall, replace)
```
